```

K&A - случаи похищения 
DD - смерти из-за приданого

```

````python
import pandas as pd
````

````python
data = pd.read_csv('CrimesOnWomenData.csv')
````

1. За какие года представлен набор данных?

````python

years = data['Year'].unique()

# Выводим года
print("Года, представленные в наборе данных:", years)
````

2. Выявите район с самым высоким количеством убийств женщин

````python
total_dd_per_state = data.groupby('State')['DD'].sum()
max_state = total_dd_per_state.idxmax()
max_dd = total_dd_per_state.max()
print(f"Штат с наибольшим количеством смертей из-за приданого: {max_state}, Количество: {max_dd}")
````

3. Какое среднее число изнасилований женщин в индии наблюдалось в указанные года?

````python
average_rape_cases = data['Rape'].mean()
print(f"Среднее количество случаев изнасилований в Индии за все года: {average_rape_cases:.2f}")
````

4. Выявите район с минимальным числом похищений за период 2001-2003 года

````python
filtered_data = data[(data['Year'] >= 2001) & (data['Year'] <= 2003)]

# Группируем данные по штатам и находим общее количество похищений за этот период
total_ka_per_state = filtered_data.groupby('State')['K&A'].sum()

# Находим штат с минимальным количеством похищений
min_state = total_ka_per_state.idxmin()
min_ka = total_ka_per_state.min()

# Вывод результата
print(f"Штат с минимальным числом похищений за период 2001-2003: {min_state}, Количество: {min_ka}")
````

5. Перечислите районы в которых наблюдались наиболее частые случае домашнего насилия за период с 2005 по 2012 года

````python

filtered_data = data[(data['Year'] >= 2005) & (data['Year'] <= 2012)]


total_dv_per_state = filtered_data.groupby('State')['DV'].sum()


top_states = total_dv_per_state.sort_values(ascending=False).head()


print("Районы с наиболее частыми случаями домашнего насилия за период 2005-2012:")
print(top_states)
````

6. Выведите названия районов в порядке не возрастания случаев торговли женщинами за 2020 год

````python

data_2020 = data[data['Year'] == 2020]

# Сортируем данные по количеству случаев торговли женщинами (WT) в порядке убывания
sorted_data = data_2020[['State', 'WT']].sort_values(by='WT', ascending=False)

# Выводим результат
print("Районы в порядке убывания случаев торговли женщинами за 2020 год:")
print(sorted_data)
````

8. Какой средний годовой доход у клиентов с уровнем лояльности больше 50 живущих на севере

````python
# Фильтруем данные по уровню лояльности > 50 и региону 'Север'
filtered_data = data[(data['loyalty_score'] > 5) & (data['region'] == 'North')]

# Вычисляем средний годовой доход среди отфильтрованных данных
average_income = filtered_data['annual_income'].mean()
# print(data['loyalty_score'].unique())
# print(data['region'].unique())
# print(f"Количество клиентов с уровнем лояльности > 50 и проживающих на Севере: {len(filtered_data)}")
# Вывод результата
print(f"Средний годовой доход у клиентов с уровнем лояльности > 50, проживающих на Севере: {average_income} долларов США")
````

9. Выведите 25 и 75 перцентили уровня лоальности клиентов живущих на западе с частотой покупок более 20 раз в год

````python
filtered_data = data[(data['region'] == 'West') & (data['purchase_frequency'] > 20)]

# Вычисляем 25-й и 75-й перцентили уровня лояльности
percentile_25 = filtered_data['loyalty_score'].quantile(0.25)
percentile_75 = filtered_data['loyalty_score'].quantile(0.75)


print(f"25-й перцентиль уровня лояльности: {percentile_25:.2f}")
print(f"75-й перцентиль уровня лояльности: {percentile_75:.2f}")
````

10. Отличается ли медианный возраст клиентов с уровнем лояльности более 70 и менее 10 для жителей Востока и Севера.

````python
east_data = data[data['region'] == 'East']
north_data = data[data['region'] == 'North']

# Медианный возраст для клиентов с уровнем лояльности > 70
median_age_east_above_70 = east_data[east_data['loyalty_score'] > 7.0]['age'].median()
median_age_north_above_70 = north_data[north_data['loyalty_score'] > 7.0]['age'].median()

# Медианный возраст для клиентов с уровнем лояльности < 10
median_age_east_below_10 = east_data[east_data['loyalty_score'] < 1.0]['age'].median()
median_age_north_below_10 = north_data[north_data['loyalty_score'] < 1.0]['age'].median()

# Вывод результатов
print(f"Медианный возраст клиентов с уровнем лояльности > 70 на Востоке: {median_age_east_above_70}")
print(f"Медианный возраст клиентов с уровнем лояльности > 70 на Севере: {median_age_north_above_70}")
print(f"Медианный возраст клиентов с уровнем лояльности < 10 на Востоке: {median_age_east_below_10}")
print(f"Медианный возраст клиентов с уровнем лояльности < 10 на Севере: {median_age_north_below_10}")

# Сравнение медианных возрастов
if median_age_east_above_70 != median_age_north_above_70:
    print("Медианный возраст клиентов с уровнем лояльности > 70 отличается между Востоком и Севером.")
else:
    print("Медианный возраст клиентов с уровнем лояльности > 70 не отличается между Востоком и Севером.")

if median_age_east_below_10 != median_age_north_below_10:
    print("Медианный возраст клиентов с уровнем лояльности < 10 отличается между Востоком и Севером.")
else:
    print("Медианный возраст клиентов с уровнем лояльности < 10 не отличается между Востоком и Севером.")
````

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Постройте диаграмму которая отражает средний уровень лояльности во всех регионах

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('CustomerPurchasingBehaviors.csv')

# Группировка данных по регионам и вычисление среднего уровня лояльности
avg_loyalty_by_region = data.groupby('region')['loyalty_score'].mean()

# Построение столбчатой диаграммы
plt.figure(figsize=(8, 6))
avg_loyalty_by_region.plot(kind='bar', color='skyblue')

# Добавление подписей и заголовков
plt.title('Средний уровень лояльности по регионам', fontsize=16)
plt.xlabel('Регион', fontsize=12)
plt.ylabel('Средний уровень лояльности', fontsize=12)
plt.xticks(rotation=0)

# Показать график
plt.show()
````

2. Постройте гистограмму распределения зарплат клиентов на Севере.

````python
north_data = data[data['region'] == 'North']

# Построение гистограммы распределения зарплат
plt.figure(figsize=(8, 6))
sns.histplot(north_data['annual_income'], bins=20, kde=True, color='skyblue')

# Добавление подписей и заголовков
plt.title('Распределение годового дохода клиентов на Севере', fontsize=16)
plt.xlabel('Годовой доход (доллары США)', fontsize=12)
plt.ylabel('Частота', fontsize=12)

# Показать график
plt.show()
````

3. Сравните уровни зарплат людей на севере и юге.

````python
# Фильтрация данных для клиентов, проживающих на Севере и Юге
north_data = data[data['region'] == 'North']
south_data = data[data['region'] == 'South']

# Построение KDE-графиков для сравнения распределений зарплат
plt.figure(figsize=(10, 6))

sns.kdeplot(north_data['annual_income'], label='Север', fill=True, color='skyblue')
sns.kdeplot(south_data['annual_income'], label='Юг', fill=True, color='salmon')

# Добавление подписей и заголовков
plt.title('Сравнение распределения годового дохода клиентов на Севере и Юге', fontsize=16)
plt.xlabel('Годовой доход (доллары США)', fontsize=12)
plt.ylabel('Плотность', fontsize=12)

# Показать легенду
plt.legend()

# Показать график
plt.show()
````

4. Постройте box plot отображающий распределение уровня лояльности для людей до 20 лет, от 20 до 40, от 40 до 60 и старше 60 лет

````python
# Создание категорий на основе возраста
bins = [0, 20, 40, 60, 100]  # Границы возрастных групп
labels = ['До 20', '20-40', '40-60', '60+']  # Названия возрастных групп
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels)

# Построение box plot для отображения распределения уровня лояльности по возрастным группам
plt.figure(figsize=(10, 6))
sns.boxplot(x='age_group', y='loyalty_score', data=data, hue='age_group', palette='Set2', legend=False)

# Добавление подписей и заголовков
plt.title('Распределение уровня лояльности по возрастным группам', fontsize=16)
plt.xlabel('Возрастные группы', fontsize=12)
plt.ylabel('Уровень лояльности', fontsize=12)

# Показать график
plt.show()
````

5. Визуализируйте зависимость частоты покупок от заработка

````python
# Построение scatter plot для отображения зависимости частоты покупок от заработка
plt.figure(figsize=(10, 6))
sns.scatterplot(x='annual_income', y='purchase_frequency', data=data, color='blue')

# Добавление подписей и заголовков
plt.title('Зависимость частоты покупок от заработка', fontsize=16)
plt.xlabel('Годовой доход (доллары США)', fontsize=12)
plt.ylabel('Частота покупок (раз в год)', fontsize=12)

# Показать график
plt.show()
````
